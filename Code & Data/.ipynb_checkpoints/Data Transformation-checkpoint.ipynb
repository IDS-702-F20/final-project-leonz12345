### Compute "is back to back"

```python
def is_b2b(row):
    team = row['Team']
    game_date = datetime.datetime.strptime(row['Date'], "%Y-%m-%d").date()
    yesterday = game_date - datetime.timedelta(days=1)
    yesterday = yesterday.strftime("%Y-%m-%d")
    if ((source['Date']==yesterday) & (source['Team']==team)).any():
        return 1
    else:
        return 0
```

```python
def is_b2b_opp(row):
    team = row['Opponent']
    game_date = datetime.datetime.strptime(row['Date'], "%Y-%m-%d").date()
    yesterday = game_date - datetime.timedelta(days=1)
    yesterday = yesterday.strftime("%Y-%m-%d")
    if ((source['Date']==yesterday) & (source['Opponent']==team)).any():
        return 1
    else:
        return 0
```

### Compute "Streak"

```python
df = {}
tm = []
sn = []
subset = []

teams = source.Team.unique()
for team in teams:
    last_game_indexes = source[(source['Game']==82) & (source['Team']==team)].index
    for i in last_game_indexes:
        season = source.loc[i-81:i]
        tm.append(season['Team'].iloc[0])
        sn.append(season['Date'].iloc[0])
        subset.append(season)

df['Team'] = tm
df['Season First Game Date'] = sn
df['Dataframe'] = subset
df = pd.DataFrame.from_dict(df)
```

```python
def compute_streak(row, games,result):
    last_game_id = row['Game']-1
    if last_game_id > 0:
        lastgame = games[games['Game']==last_game_id]
        prev_strk = result[last_game_id-1]
        prev_w_l = lastgame.WINorLOSS.item()
        if prev_strk >= 0 and prev_w_l == 'W':
            result.append(prev_strk+1)
            #row['Streak'].iat[44] = prev_strk+1
            #game.iat[last_game_id,44] = prev_strk+1
        elif prev_strk <= 0 and prev_w_l == 'L':
            result.append(prev_strk-1)
        elif prev_w_l == 'W':
            result.append(1)
        else:
            result.append(-1)
            #row['Streak'].iat[44] = prev_strk-1
            #game.iat[last_game_id,44] = prev_strk-1
```
